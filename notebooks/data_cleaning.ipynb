### Task 8. Outliers in the Vehicle's Year of Manufacture

1. In this step, I display the descriptive statistics for the vehicle's year of manufacture and pay attention to the boundary values and quartiles.



```python
df['year'].describe()
```

2. In this step, I calculate the boundaries for normal values of the vehicle's year of manufacture using the interquartile range (IQR). I then examine whether the minimum and maximum values fall within the interval defined by these boundaries.


```python
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    # Calculation of the Interquartile Range (IQR)
    igr = q75 - q25
    # Outlier Boundaries
    boundaries = (q25 - 1.5 * igr, q75 + 1.5 * igr)
    return boundaries
```

```python
 boundaries = calculate_outliers(df['year'])
boundaries
```

```python
print(df['year'].min(), df['year'].max())
print(boundaries)
```

3. In this step, I calculate and display the number of outliers in the vehicle's year of manufacture.

```python
lower, upper = boundaries
outliers = df[(df['year'] < lower) | (df['year'] > upper)]
len(outliers)
```

4. In this step, I calculate and display the proportion of outliers in the vehicle's year of manufacture.


```python
outliers_percentage = len(outliers) / len(df)
print('Proportion of Outliers in the Vehicles Year of Manufacture: ', outliers_percentage, '%')
```

5. In this step, I display the records where the vehicle's year of manufacture falls into the outlier range. I then examine the data to determine whether these values represent specific price categories or other features, or if they are truly outliers.

```python
outliers
```

6. In this step, I display the number of values for the price category among the records where the vehicle's year of manufacture is considered an outlier. I then examine whether these anomalous year values describe any specific price category better than the others.


```python
outliers['price_category'].value_counts()
```

7. Since the outliers in the vehicle's year of manufacture are among the minimum values, I replace those outlier values with the lower boundary. I also make sure that the year remains an integer value within the range of normal values.

```python
import math
lower_int = math.ceil(lower)
df.loc[df['year'] < lower, 'year'] = lower_int
```

```python
df['year'].describe()
```
